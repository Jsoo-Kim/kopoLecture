## 데이터 조작하기 (np.where 활용)

```python
dataUrl = "https://raw.githubusercontent.com/hyokwan/python-lecture/master/dataset/customerdata.csv"
customerData = pd.read_csv(dataUrl)
```

```python
# 새로운 컬럼 생성
customerData["PRODUCTAGE_NEW999"] = "$"
# PRODUCTAGE가 1보다 작으면 1 아니면 2
np.where( customerData.PRODUCTAGE < 1, 1, 2)
```

```python
customerData
```

```python
# 컬럼 삭제 - 할당 안 해서 저장 안 됨
customerData.drop(columns=["PRODUCTAGE_NEW999"])
customerData
```

```python
# 컬럼 삭제 - 할당해서 저장 됨
customerData = customerData.drop(columns=["PRODUCTAGE_NEW999"])
customerData
```

```python
np.where( "조건", True, False )
np.where( "조건", True, np.where( "조건", True, False ) )
```

```python
customerData["PRODUCTAGENEW3"] = np.where( customerData.PRODUCTAGE < 1, 1,
         np.where( customerData.PRODUCTAGE < 2, 2, 3 ) )
customerData
```

```python
### 데이터 조작 #1 (np.where : 간단한 데이터 조작 시 속도 빠름)
### 데이터 조작 #2 (함수를 활용한 조작 : 로직이 복잡할 때)
```

```python
def productAgePreProcessing(inValue, addValue=0):
    """
        docstring
        description: 함수 설명
        param: 
        return:
        ...
    """
    # inValue = 1.2
    outValue = 0
    
    if inValue < 1:
        outValue = 1
    elif inValue < 2:
        outValue = 2
    elif inValue < 3:
        outValue = 3
    else:
        outValue = 5
    outValue += addValue
    return outValue
```

```python
productAgePreProcessing(1.7, 4)
productAgePreProcessing(1.7)
```

```python
customerData["PRODUCTAGE_NEW4"] = customerData.PRODUCTAGE.apply( productAgePreProcessing )
customerData
```

## 데이터 조작하기 (loc)

```python
customerData.loc[ customerData.PRODUCTAGE < 1, "PRODUCTAGE_LOC" ] = 1  # 조건에 해당하는 "PRODUCTAGE_LOC" 컬럼의 값을 1로 업데이트
customerData.loc[ (customerData.PRODUCTAGE >= 1) & 
                  (customerData.PRODUCTAGE < 2), "PRODUCTAGE_LOC" ] = 2
customerData.loc[ customerData.PRODUCTAGE < 3, "PRODUCTAGE_LOC" ] = 3
customerData.loc[ customerData.PRODUCTAGE >= 3, "PRODUCTAGE_LOC" ] = 4
customerData
```
