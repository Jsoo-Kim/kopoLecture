```python
import pandas as pd
import numpy as np
```

```python
selloutData = pd.read_csv("../dataset/kopo_channel_seasonality_new.csv")
selloutData
```

```python
selloutData["QTY_NEW"] = np.where( selloutData.QTY < 0, 0, selloutData.QTY )
selloutData
# selloutData.loc[ selloutData.QTY < 0 ]
```

```python
selloutData["YEAR"] = selloutData.YEARWEEK.astype(str).str[:4]
selloutData["WEEK"] = selloutData.YEARWEEK.astype(str).str[4:]

refinedSelloutData = selloutData.loc[ selloutData.WEEK.astype(int) <= 52 ]
refinedSelloutData
```

```python
sortKey = ["REGIONID", "PRODUCT", "YEARWEEK"]
```

```python
sortedData = refinedSelloutData.sort_values(by=sortKey, ascending=[True, True, False]) # 어차피 다 오름차순이니까 ascending은 생략 가능
sortedData
```

```python
groupKey = ["REGIONID", "PRODUCT", "YEAR"]
```

```python
groupData = sortedData.groupby(by=groupKey)[["QTY_NEW"]].agg("mean").reset_index(drop=False)
# .reset_index(drop=False)
groupData = groupData.rename( columns={ "QTY_NEW":"QTY_MEAN" })
groupData
```
