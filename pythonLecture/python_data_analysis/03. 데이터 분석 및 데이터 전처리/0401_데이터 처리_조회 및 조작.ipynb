## 원하는 행 조회하기

```python
emiCondition = 2
```

```python
emi3csData = customerData.loc[ customerData.EMI > emiCondition ] # [] 안에 조건! 비교할 때 타입도 일치해야 함
emi3csData
```

```python
emi3csData.shape
```

```python
customerData.loc[ 0:4 ] # 원래 없던 인덱스가 맨 왼쪽에 생기면서 조회! loc은 특이하게 첫 인덱스부터 '마지막 인덱스 전까지'가 아니라 '마지막 인덱스까지(4까지)' 조회함
```

```python
## avgPrice > 2000
## emi > 2
```

```python
avgPriceCond = 2000
emiCond = 2
```

```python
# customerData.loc[ customerData.AVGPRICE > avgPriceCond & customerData.EMI > emiCond ] # 이렇게 하면 에러
pr2000emi3Data = customerData.loc[ (customerData.AVGPRICE > avgPriceCond) & (customerData.EMI > emiCond) ] # 이렇게 각 조건에 괄호 넣어줘야 함
```
