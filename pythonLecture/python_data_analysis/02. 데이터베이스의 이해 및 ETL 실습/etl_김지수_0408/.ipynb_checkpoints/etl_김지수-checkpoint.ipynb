```python
# postgresql 용 라이브러리
import psycopg2
# 행과열 처리용 라이브러리
import pandas as pd
# db 접속 엔진 생성용 라이브러리
from sqlalchemy import create_engine
# MariaDB 또는 MySQL 데이터베이스에 연결하고 상호작용하기 위한 패키지
import pymysql
```

```python
# 데이터 받아올 DB 접속 정보 
dbPrefix = "postgresql"
dbId = "postgres"
dbPw = "postgres"
dbIp = "192.168.110.111"
dbPort = "5432"
dbName = "postgres"

pgEngineSchool = create_engine(f"{dbPrefix}://{dbId}:{dbPw}@{dbIp}:{dbPort}/{dbName}")
```

```python
# 테이블 생성해서 내보낼 외부 DB 접속 정보 
dbPrefix = "mysql+pymysql"
dbId = "hkcode"
dbPw = "hkcodemariadb"
dbIp = "139.150.80.119"
dbPort = "3306"
dbName = "hkcodedb"

pgEngineOut = create_engine(f"{dbPrefix}://{dbId}:{dbPw}@{dbIp}:{dbPort}/{dbName}")
```

```python
# 데이터 받아올 DB에 날릴 쿼리
selSelect = """
select *
from csdata_kopo
"""
```

```python
# 데이터 받아올 DB에서 데이터 추출
inData = pd.read_sql_query(selSelect, pgEngineSchool)
inData
```

```python
# 조건에 해당되는 데이터만 추출
emiCondition = 2
avgPrice = 3200

cumstomedData = inData.loc[ (inData.avgprice >= avgPrice) &
                (inData.emi == emiCondition) ]
cumstomedData
```

```python
# 조건에 맞는 데이터만 담은 테이블 생성해서 외부 DB에 넣어줌
tableName = "csdata_김지수"
cumstomedData.to_sql(name = tableName,
              con = pgEngineOut,
              if_exists = "replace", 
              index=False)
```
